```python
import numpy as np
import pandas as pd
import cv2 
from google.colab.patches import cv2_imshow 
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```python
Image=cv2.imread("sis.jpg")
cv2_imshow(Image)
```

```python
image=cv2.imread("sis.jpg",0)
cv2_imshow(image)
```

```python
def conv_transform(image):
  image_copy=image.copy()
  for i in range(image.shape[0]):
    for j in range(image.shape[1]):
      image_copy[i][j]=image[image.shape[0]-i-1][image.shape[1]-j-1]
  return image_copy
```

```python
def conv(image,kernel):
  kernel= conv_transform(kernel)
  image_h=image.shape[0]
  image_w=image.shape[1]
  
  kernel_h=kernel.shape[0]
  kernel_w=kernel.shape[1]
  
  h=kernel_h//2
  w=kernel_w//2

  image_conv=np.zeros(image.shape)
  for i in range(h,image_h-h):
    for j in range(w,image_w-w):
      sum=0

      for m in range(kernel_h):
        for n in range(kernel_w):
          sum=sum+kernel[m][n]*image[i-h+m][j-w+n]

      image_conv[i][j]=sum
  return image_conv

  conv(image,kernel)
```

```python
def norm(img1,img2):
  img_copy=np.zeros(img1.shape)
  for i in range(img1.shape[0]):
    for j in range(img1.shape[1]):
      q=(img1[i][j]**2+img2[i][j]**2)**(1/2)
      if(q>6):
        img_copy[i][j]=0
      else:
        img_copy[i][j]=255
  return img_copy
print(norm)
```

```python
kernel=np.zeros(shape=(3,3))
kernel[0,0]=-1
kernel[0,1]=-2
kernel[0,2]=-1
kernel[1,0]=0
kernel[1,1]=0
kernel[1,2]=0
kernel[2,0]=1
kernel[2,1]=2
kernel[2,2]=1
gy=conv(image,kernel)
cv2_imshow(gy)





```

```python
kernel[0,0]=-1
kernel[0,1]=0
kernel[0,2]=1
kernel[1,0]=-1
kernel[1,1]=0
kernel[1,2]=1
kernel[2,0]=-2
kernel[2,1]=0
kernel[2,2]=2
gx=conv(image,kernel)
cv2_imshow(gx)
g_sobel=norm(gx,gy)
```

```python
cv2_imshow(g_sobel)
```

```python
kernel2=np.ones((3,3),np.uint8)
```

```python
erosion=cv2.erode(g_sobel,kernel2,iterations=1)
cv2_imshow(erosion)
```

```python
kernel3=np.ones((4,4),np.uint8)
```

```python
dilation=cv2.dilate(erosion,kernel3,iterations=1)

cv2_imshow(dilation)
```

```python
img3 = dilation.astype(np.uint8)
print(img3.dtype)
print(img3.shape)
```

```python
lab_image=cv2.cvtColor(Image,cv2.COLOR_BGR2LAB)
cv2_imshow(lab_image)
```

```python
print(lab_image.shape)
```

```python
print(lab_image.shape)
```

```python
l,a,b=cv2.split(lab_image)
```

```python
plt.hist(l.flat,bins=100,range=(0,255))
```

```python
equ=cv2.equalizeHist(l)
```

```python
plt.hist(equ.flat,bins=100,range=(0,255))
```

```python
updated_lab_image=cv2.merge((equ,a,b))
```

```python
hist_equ_img=cv2.cvtColor(updated_lab_image,cv2.COLOR_LAB2BGR)
cv2_imshow(hist_equ_img)
```

```python
clahe=cv2.createCLAHE(clipLimit=3.0,tileGridSize=(25,25))
```

```python
clahe_image=clahe.apply(l)
```

```python
cv2_imshow(clahe_image)
```

```python
plt.hist(clahe_image.flat,bins=100,range=(0,255))
```

```python
lab_image2=cv2.merge((clahe_image,a,b))
```

```python
CLAHE_image=cv2.cvtColor(lab_image2,cv2.COLOR_LAB2BGR)

cv2_imshow(CLAHE_image)
```

```python
son=cv2.bitwise_and(CLAHE_image,CLAHE_image,img3)
cv2_imshow(son)
```

```python
filtre=cv2.blur(son,(1,1))
cv2_imshow(filtre)
```
